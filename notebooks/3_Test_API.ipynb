```python
import requests
import pandas as pd
import json
from datetime import datetime
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Transform kelvin to celcius
def kelvin_cel(temp_k):
    return temp_k - 273.15
```

```python
# Get the data from our city
api_k = '746d0f931915f114220df0290d885fbc'
city = 'Ibagu√©,CO'
url = f'http://api.openweathermap.org/data/2.5/forecast?q={city}&cnt=240&appid={api_k}' #240 hours

response = requests.get(url)
```

```python
if response.status_code == 200:
    data = response.json()
    weather_data = []
    for hour in data['list']:
        date_time = datetime.utcfromtimestamp(hour['dt']).strftime('%Y-%m-%d %H:%M:%S')
        temp = kelvin_cel(hour['main']['temp'])
        temp_min = kelvin_cel(hour['main']['temp_min'])
        temp_max = kelvin_cel(hour['main']['temp_max'])
        pressure = hour['main']['pressure']
        humidity = hour['main']['humidity']

        weather_main = hour['weather'][0]['main']
        weather_description = hour['weather'][0]['description']
        wind_speed = hour['wind']['speed']
        rain_volume = hour.get('rain',{}).get('3h',0)

        weather_data.append({
        'date_time' : date_time,
        'temp': temp,
        'temp_min': temp_min,
        'temp_max': temp_max,
        'pressure': pressure,
        'humidity': humidity,
        'weather_main': weather_main,
        'weather_description': weather_description,
        'wind_speed': wind_speed,
        'rain_volume': rain_volume
        })
    df = pd.DataFrame(weather_data)
    df['sub_zona_id'] = 13
else:
    print(f'Failed to get data: {response.status_code}')
```
